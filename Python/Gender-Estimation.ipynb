# Gender Estimation for First Names

```python
FemaleNames = []
with open('names-female.txt', 'r') as file:
    for name in file:
        FemaleNames.append(name[:-1]) # discard newline
```

```python
FemaleNames
```

```python
MaleNames = []
with open('names-male.txt', 'r') as file:
    for name in file:
        MaleNames.append(name[:-1]) # discard newline
MaleNames
```

```python
from string import ascii_lowercase as alphabet
alphabet
```

Let us compute the prior probabilities $P(\texttt{Female})$ and $P(\texttt{Male})$.

```python
pFemale = len(FemaleNames) / (len(FemaleNames) + len(MaleNames))
pMale   = len(MaleNames)   / (len(FemaleNames) + len(MaleNames))
```

```python
def conditional_prop(c, g):
    if g == 'f':
        return len([n for n in FemaleNames if n[-1] == c]) / len(FemaleNames)
    else:
        return len([n for n in MaleNames   if n[-1] == c]) / len(MaleNames)
```

```python
Conditional_Probability = {}
for c in alphabet:
    Conditional_Probability[(c, 'f')] = conditional_prop(c, 'f')
    Conditional_Probability[(c, 'm')] = conditional_prop(c, 'm')
```

```python
def estimate(name):
    last   = name[-1]
    female = Conditional_Probability[(last, 'f')] / pFemale
    male   = Conditional_Probability[(last, 'm')] / pMale
    if female >= male:
        return 'f'
    else:
        return 'm'
```

```python
estimate('Karl')
```

```python
estimate('Jessica')
```

```python
total   = 0
correct = 0
for n in FemaleNames:
    if estimate(n) == 'f':
        correct += 1
    total += 1
for n in MaleNames:
    if estimate(n) == 'm':
        correct += 1
    total += 1
accuracy = correct / total
accuracy
```
